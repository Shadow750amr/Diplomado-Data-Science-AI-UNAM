## 1) Fundamentos y tipos básicos

**1.1** Crea una función `area_rectangulo(base, altura)` que regrese el área. Valida que `base` y `altura` sean números positivos; si no, regresa `None`.

```python
def area_rectangulo(base, altura):
    # Tu código aquí

    if base > 0 and altura > 0: # valida que base y altura sean positivos
      return base * altura #retorna la multiplicación
    else:
      return None
    pass

# Pruebas sugeridas
print(area_rectangulo(3, 4))  # 12
print(area_rectangulo(-1, 2)) # None
```

**Explicación:** En este caso simplemente se agregó una condición if dentro de la función para validar que tanto base como altura fueran positivos y luego retornar la multiplicación si es que se cumpliera esta condición. Lo siguiente fue simplemente retornar None en caso de que esto fuera falso.

**1.2** Escribe una función `es_par(n)` que regrese `True` si `n` es par y `False` en caso contrario.

```python
def es_par(n):
    # Tu código aquí
    if n % 2 == 0:  # valida que el residuo sea igual cero
      return True  # y retorna verdadero
    else:
      return False
    pass

print(es_par(4))  # True
print(es_par(7))  # False
```

**Explicación:** Lo que hice fue agregar una condición if dentro de la función para validar que el residuo de dividir 'n' entre 2 fuera cero, para así poder saber qué numero es par o impar y así retornar true o false.

## 2) Condicionales y ciclos

**2.1** Dada una lista de números, crea `cuenta_mayores(lista, umbral)` que cuente cuántos son estrictamente mayores que `umbral`.

```python
def cuenta_mayores(lista, umbral):
    # Tu código aquí
    contador = 0
    for numero in lista:
      if numero > umbral:   #valida si el numero de la lista es mayor al umbral (3)
        contador += 1       #suma el valor adicional cuando cumple la condición
    return contador
    pass

print(cuenta_mayores([1,5,3,10,2], 3))  # 3
```

**Explicación:** Lo que hice fue agregar primero un contador que "almacena" cuantas veces se repite la condición y empieza en cero. Luego, agregué un ciclo for con un if statement que itera para cada numero en la lista, si el numero es mayor al umbral definido en la función, en este caso 3, le suma un numero al contador. Así se llega al resultado de que únicamente 5 y 10 son mayores que 3, es decir, dos numeros de la lista cumplen la condición.

**2.2** Imprime los primeros 10 números de la secuencia de Fibonacci usando un ciclo `for` o `while`.

```python
def fibonacci(n):
    a, b = 0, 1     #Definimos a y b
    for _ in range(n):
        a, b = b, a + b  # Para cada elemento de la lista, mantiene el valor anterior y le suma el nuevo valor
    return a

lista = []
for i in range(10):
    lista.append(fibonacci(i))

print(lista)

```

**Explicación:** Lo que esto es crea es una función llamada fibonacci con el argumento 'n'. Dentro de esta función se establecen las variables a y b, que son 0 y 1, la función usa esos números para iterar y sumar el numero anterior con el nuevo numero de la lista. Luego simplemente usamos una lista para almacenar el resultado de lo que se aplique con el ciclo for i in range, que es el numero de iteraciones que buscamos y lo almacenamos en la lista que creamos con el append conviertiendo a string la funcion de fibonacci. Al final imprimimos el resultado.



## 3) Listas, tuplas, conjuntos y diccionarios

**3.1** Dada una lista de palabras, crea `palabras_unicas(lista)` que regrese un **conjunto** con las palabras únicas en minúsculas.

```python
def palabras_unicas(lista):
    for i in range(len(lista)):  #Definimos el rango de la lista y usamos for i para iterar en cada elemento
        lista[i] = lista[i].lower()  # Convertimos a minuscula cada elemento
    return set(lista)  # Y traemos el valor unico
    pass

print(palabras_unicas(['Hola','mundo','hola','Python']))
```

**Explicación:** Esta función itera en una lista y sustituye el valor de esa lista por el mismo valor en minusculas, luego retorna los valores unicos de esta lista con set.

**3.2** Construye `conteo_letras(cadena)` que cuente cuántas veces aparece cada letra (solo a‑z, ignora espacios y signos) y regrese un **diccionario**.

```python
def conteo_letras(cadena):
    # Tu código aquí
    alfabeto = 'abcdefghijklmnopqrstuvwxyz'  #Definimos el alfabeto
    diccionario = {}     # y el diccionario
    for letra in cadena.lower():   #para cada letra en la cadena de texto
      if letra in alfabeto:
        diccionario[letra] = diccionario.get(letra, 0) + 1  #traemos la letra y le añadimos el contador
    return diccionario

print(conteo_letras('Hola mundo'))
```

**Explicación:** Este código define un diccionario (donde se almacenan los resultados) y un alfabeto. El ciclo for valida si la letra de la cadena está en el alfabeto y devuelve un diccionario con el contador de veces que se repite la letra.

## 4) Funciones (simples)

**4.1** Escribe `promedio(lista)` que regrese el promedio de una lista de números (si la lista está vacía, regresa `None`).

```python
import numpy as np
def promedio(lista):
    lista = np.array(lista)  #Definimos la lista
    if lista.size == 0:      #Si la lista esta vacia
      return None             # retorna None
    else:
      return lista.mean()     # Si no, retorna el promedio
    pass

print(promedio([10, 8, 9]))
print(promedio([]))
```

**Explicación:** usé numpy para pasar la lista a un array y poder promediarla. Básicamente integré un if para validar si la lista está vacía y poder sacar el promedio en caso de que no esté vacía.

**6.1** Dada una **lista de listas** con números (por ejemplo, una "matriz"), crea `suma_filas(matriz)` que regrese una lista con la **suma** de cada fila.

```python
import numpy as np
def suma_filas(matriz):
    matriz = np.array(matriz)    #Definimos la matriz
    return matriz.sum(axis=1)   # y aplicamos la función sum
    pass

print(suma_filas([[1,2,3],[4,5,6],[7,8,9]]))  # [6, 15, 24]
```

**Explicación:** Aquí se almacenan números en un array usando numpy. Luego uso el método sum para sumar los valores de esa matriz por fila de forma vectorizada.

## 7) Manejo básico de errores

**7.1** Implementa `division_segura(a,b)` que divida `a/b` y capture `ZeroDivisionError`, regresando `None` en ese caso.

```python
def division_segura(a,b):
    # Tu código aquí
    try:
        return a / b   #retorna a/b
    except ZeroDivisionError:   #pero si hay error retorna None
        return None

print(division_segura(10,2))
print(division_segura(10,0))
```

**Explicación:** Usé un try para capturar el valor de a/b y luego aplique un except que devuelve None en caso de que la división sea incorrecta.
